```python
# Dependencies
import pandas as pd
import numpy as np

# Files to load
schools_file_csv = "schools_complete.csv"
students_file_csv = "students_complete.csv"

# Files to read
schools_file = pd.read_csv(schools_file_csv)
students_file = pd.read_csv(students_file_csv)

```

```python
schools_file.head()
```

```python
total_students = schools_file["size"].sum()
total_schools = schools_file["School ID"].count()
total_budget = schools_file["budget"].sum()
print(total_students)
print(total_schools)
print(total_budget)
```

```python
students_file.head()
```

```python
average_math = round(students_file["math_score"].mean(),2)
average_reading = round(students_file["reading_score"].mean(),2)
passing_rate = (average_math + average_reading)/2

print(average_math)
print(average_reading)
print(passing_rate)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
passing_math = students_file.loc[students_file['math_score'] >= 70]
passing_math.head()

```

```python
total_pass_math = passing_math["math_score"].count()
print(total_pass_math)

percent_pass_math = round((total_pass_math/total_students)*100,2)
print(percent_pass_math)
```

```python
passing_read = students_file.loc[students_file['reading_score'] >= 70]
passing_read.head()
```

```python
total_pass_read = passing_read["reading_score"].count()
print(total_pass_read)


percent_pass_read = round((total_pass_read/total_students)*100,2)
print(percent_pass_read)
```

```python
District_summary = pd.DataFrame({"Total Schools":[total_schools],
                           "Total Students":[total_students],
                           "Total Budget":[total_budget],
                           "Average Math Score": [average_math],
                           "Average Reading Score": [average_reading],
                            "% Passing Math" : [percent_pass_math],
                           "% Passing Reading":[percent_pass_read],
                            "Overall Passing Rate": [passing_rate]})
District_summary["Total Students"] = District_summary["Total Students"].map("{:,}".format)
District_summary["Total Budget"] = District_summary["Total Budget"].map("${:,.2f}".format)
District_summary
```

```python
# Combine data into one dataset
school_complete = pd.merge(schools_file, students_file, on="school_name")
school_complete.head()
```

```python
del school_complete['School ID']
```

```python
del school_complete['student_name']
```

```python
del school_complete['gender']
```

```python
school_complete.head()
```

```python
#School Summary

schools = schools_file["school_name"]
schools
```

```python
#School Type
school_type = schools_file["type"]
school_type
```

```python
#Total Students
total_students = school_complete.groupby("school_name")["Student ID"].count()
print(total_students)
```

```python
#Total School Budget
school_budget = schools_file.groupby("school_name")["budget"].sum()
print(school_budget)
```

```python
#Per Student Budget
per_student_budget = school_budget/total_students
print(per_student_budget)
```

```python
#Average Math Score

avg_math = school_complete.groupby("school_name")["math_score"].mean()
avg_math
```

```python
#Average Reading Score
avg_read = school_complete.groupby("school_name")["reading_score"].mean()
avg_read
```

```python
#% Passing Math
pass_math = round(students_file.loc[students_file['math_score'] >= 70].groupby("school_name")["Student ID"].count()/total_students*100,2)
print(pass_math)
```

```python
#% Passing Reading
pass_read = round(students_file.loc[students_file['reading_score'] >= 70].groupby("school_name")["Student ID"].count()/total_students*100,2)
print(pass_read)
```

```python
#Overall Passing Rate (Average of the above two)
overall_pass_rate = (pass_math + pass_read)/2
print(overall_pass_rate)
```

```python
School_Summary = pd.DataFrame({"School Type" : school_type,
                               "Total Students" : total_students,
                               "Total School Budget" : school_budget,
                               "Budget Per Student" : per_student_budget,
                               "Average Math Score" : avg_math,
                               "Average Reading Score" : avg_read,
                               "% Passing Math" : pass_math,
                               "% Passing Reading" : pass_read,
                               "Overall Passing Rate" : overall_pass_rate})
School_Summary["Total Students"] = School_Summary["Total Students"].map("{:,}".format)
School_Summary["Total School Budget"] = School_Summary["Total School Budget"].map("${:,.2f}".format)
School_Summary
```

```python
#Top Performing Schools (By Passing Rate) max
top_performing = School_Summary.sort_values('Overall Passing Rate', ascending=False)
top_performing.head()
```

```python
#Bottom Performing Schools (By Passing Rate) min
bottom_performing = School_Summary.sort_values('Overall Passing Rate')
bottom_performing.head()
```

```python
#Math Scores by Grade
math_scores_by_grade = round(students_file.pivot_table(index="school_name", columns="grade", values="math_score"),2)
math_scores_by_grade
```

```python
#Reading Scores by Grade
reading_scores_by_grade = round(students_file.pivot_table(index="school_name", columns="grade", values="reading_score"),2)
reading_scores_by_grade
```

```python
#Scores by School Spending
scores_by_school_spending = pd.DataFrame({"Budget Per Student" : per_student_budget,
                                        "Average Math Score": avg_math,
                                         "Average Reading Sore": avg_read,
                                         "% Passing Math": pass_math,
                                        "% Passing Reading": pass_read,
                                        "% Overall Passing rate":overall_pass_rate})


scores_by_school_spending= scores_by_school_spending.set_index("Budget Per Student")   
```

```python
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

School_Summary["Budget Ranges"] = pd.cut(per_student_budget, spending_bins, labels=group_names)
```

```python
spending_math_scores = School_Summary.groupby(["Budget Ranges"]).mean()["Average Math Score"]
spending_reading_scores = School_Summary.groupby(["Budget Ranges"]).mean()["Average Reading Score"]
spending_passing_math = School_Summary.groupby(["Budget Ranges"]).mean()["% Passing Math"]
spending_passing_reading = School_Summary.groupby(["Budget Ranges"]).mean()["% Passing Reading"]
overall_passing_rate = (spending_passing_math + spending_passing_reading) / 2
```

```python
school_spending = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})
school_spending
```

```python
#Scores by School Size
size_bins = [0, 2000, 4000, 6000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"] 

School_Summary["Size Ranges"] = pd.cut(total_students, size_bins, labels=group_names)
```

```python
size_math_scores = School_Summary.groupby("Size Ranges").mean()["Average Math Score"]
size_reading_scores = School_Summary.groupby("Size Ranges").mean()["Average Reading Score"]
size_passing_math = School_Summary.groupby("Size Ranges").mean()["% Passing Math"]
size_passing_reading = School_Summary.groupby("Size Ranges").mean()["% Passing Reading"]
size_passing_rate = (size_passing_math + size_passing_reading) / 2
```

```python
scores_by_school_size = pd.DataFrame({"Average Math Score": size_math_scores,
                                         "Average Reading Sore": size_reading_scores,
                                         "% Passing Math": size_passing_math,
                                        "% Passing Reading": size_passing_reading,
                                        "% Overall Passing rate":size_passing_rate})


scores_by_school_size   
```
